```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
file_path = r'C:\Users\anton\OneDrive\Documentos\GitHub\DataScience\data\ShoreGroveAuto.xlsx'
data = pd.read_excel(file_path, sheet_name='Shore Grove Auto')

# Adicionar coluna de lucro
data['Profit'] = data['Sale Price'] - data['Buy Price']

# Gráfico 1: Lucro total por concessionária
total_profit = data.groupby('Dealership Name')['Profit'].sum()
plt.figure(figsize=(10, 5))
total_profit.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Lucro Total por Concessionária', fontsize=16)
plt.xlabel('Concessionária', fontsize=12)
plt.ylabel('Lucro Total (em $)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Gráfico 2: Média de lucro por concessionária
average_profit = data.groupby('Dealership Name')['Profit'].mean()
plt.figure(figsize=(10, 5))
average_profit.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Média de Lucro por Concessionária', fontsize=16)
plt.xlabel('Concessionária', fontsize=12)
plt.ylabel('Média de Lucro (em $)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Configurar o tamanho do gráfico
plt.figure(figsize=(12, 6))

# Criar o gráfico de dispersão
plt.scatter(data['Sale Price'], data['Profit'], color='green', alpha=0.7, edgecolor='k')

# Personalizar o gráfico
plt.title('Relação entre Lucro e Preço de Venda', fontsize=16)
plt.xlabel('Preço de Venda (em $)', fontsize=12)
plt.ylabel('Lucro (em $)', fontsize=12)
plt.grid(alpha=0.5, linestyle='--')

# Mostrar o gráfico
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Contar o número de vendas por concessionária
sales_count = data.groupby('Dealership Name')['Sale ID'].count()

# Calcular o lucro total por concessionária
total_profit_per_dealer = data.groupby('Dealership Name')['Profit'].sum()

# Configurar o gráfico
fig, ax1 = plt.subplots(figsize=(12, 6))

# Gráfico de barras para o número de vendas
ax1.bar(sales_count.index, sales_count, color='lightblue', edgecolor='black', label='Número de Vendas')

# Criar um eixo secundário para o lucro
ax2 = ax1.twinx()
ax2.plot(total_profit_per_dealer.index, total_profit_per_dealer, color='orange', marker='o', label='Lucro Total', linewidth=2)

# Personalizar o gráfico
ax1.set_xlabel('Concessionária', fontsize=12)
ax1.set_ylabel('Número de Vendas', fontsize=12)
ax2.set_ylabel('Lucro Total (em $)', fontsize=12)
ax1.set_title('Lucro e Número de Vendas por Concessionária', fontsize=16)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar a legenda
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Ajustar a exibição
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
