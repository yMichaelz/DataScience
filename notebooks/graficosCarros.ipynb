```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar os dados
file_path = r'C:\Users\anton\OneDrive\Documentos\GitHub\DataScience\data\ShoreGroveAuto.xlsx'
data = pd.read_excel(file_path, sheet_name='Shore Grove Auto')

# Adicionar coluna de lucro
data['Profit'] = data['Sale Price'] - data['Buy Price']

# Grfico 1: Lucro total por concessionria
total_profit = data.groupby('Dealership Name')['Profit'].sum()
plt.figure(figsize=(10, 5))
total_profit.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Lucro Total por Concessionria', fontsize=16)
plt.xlabel('Concessionria', fontsize=12)
plt.ylabel('Lucro Total (em $)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Grfico 2: Mdia de lucro por concessionria
average_profit = data.groupby('Dealership Name')['Profit'].mean()
plt.figure(figsize=(10, 5))
average_profit.plot(kind='bar', color='orange', edgecolor='black')
plt.title('Mdia de Lucro por Concessionria', fontsize=16)
plt.xlabel('Concessionria', fontsize=12)
plt.ylabel('Mdia de Lucro (em $)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Configurar o tamanho do grfico
plt.figure(figsize=(12, 6))

# Criar o grfico de disperso
plt.scatter(data['Sale Price'], data['Profit'], color='green', alpha=0.7, edgecolor='k')

# Personalizar o grfico
plt.title('Relao entre Lucro e Preo de Venda', fontsize=16)
plt.xlabel('Preo de Venda (em $)', fontsize=12)
plt.ylabel('Lucro (em $)', fontsize=12)
plt.grid(alpha=0.5, linestyle='--')

# Mostrar o grfico
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt

# Contar o nmero de vendas por concessionria
sales_count = data.groupby('Dealership Name')['Sale ID'].count()

# Calcular o lucro total por concessionria
total_profit_per_dealer = data.groupby('Dealership Name')['Profit'].sum()

# Configurar o grfico
fig, ax1 = plt.subplots(figsize=(12, 6))

# Grfico de barras para o nmero de vendas
ax1.bar(sales_count.index, sales_count, color='lightblue', edgecolor='black', label='Nmero de Vendas')

# Criar um eixo secundrio para o lucro
ax2 = ax1.twinx()
ax2.plot(total_profit_per_dealer.index, total_profit_per_dealer, color='orange', marker='o', label='Lucro Total', linewidth=2)

# Personalizar o grfico
ax1.set_xlabel('Concessionria', fontsize=12)
ax1.set_ylabel('Nmero de Vendas', fontsize=12)
ax2.set_ylabel('Lucro Total (em $)', fontsize=12)
ax1.set_title('Lucro e Nmero de Vendas por Concessionria', fontsize=16)
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar a legenda
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Ajustar a exibio
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
